```python
import pandas as pd
from IPython.display import display, HTML

chemin_fichier = 'resultats-par-niveau-burvot-t1-france-entiere.xlsx'
file = pd.read_excel(chemin_fichier, engine='openpyxl')
file.to_csv('fichier')
```

```python
# Join all column names into a single string and print
columns_string = ', '.join(file.columns)
print(columns_string)
```

```python
candidates = [
    "Nathalie_ARTHAUD", "Fabien_ROUSSEL", "Emmanuel_MACRON", "Jean_LASSALLE",
    "Marine_LE_PEN", "Eric_ZEMMOUR", "Jean-Luc_MÉLENCHON", "Anne_HIDALGO",
    "Yannick_JADOT", "Valérie_PÉCRESSE", "Philippe_POUTOU", "Nicolas_DUPONT-AIGNAN"
]

def rename_columns(df):
    new_names = {}
    columns_to_drop = []
    
    for i, candidate in enumerate(candidates):
        base_index = 21 + i * 7
        
        if base_index + 6 < len(df.columns):
            new_names.update({
                df.columns[base_index + 2]: f'candidat_{candidate}',
                df.columns[base_index + 5]: f'voix/ins_{candidate}',
                df.columns[base_index + 6]: f'voix/exp_{candidate}'
            })
            
    for i, candidate in enumerate(candidates):
        base_index = 21 + i * 7
        
        if base_index + 6 < len(df.columns):
            columns_to_drop.extend([df.columns[base_index], df.columns[base_index + 1], df.columns[base_index + 3], df.columns[base_index + 4]])        
    
    # Rename columns
    df = df.rename(columns=new_names)
    
    # Drop unnecessary columns
    df = df.drop(columns=columns_to_drop)
    
    return df


df_processed = rename_columns(file)
```
